# Ridge Maps

Uses the module developed here: https://github.com/ColCarroll/ridge_map

Requires this module also: https://github.com/tkrajina/srtm.py

A useful way to get the bounding box of the plotting area easily: http://bboxfinder.com/

Import standard modules:

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.colors as colors

import sys
sys.path.append('../')
from functions import *
```

Import the `ridge_map` and `srtm` modules:

```python
import sys
sys.path.append('../../../python_pkgs/ridge_map/ridge_map/')
sys.path.append('../../../python_pkgs/srtm/')
from ridge_map import *
```

Create our own colormap:

```python
c1 = '#112e40'
c2 = '#175073'
c3 = '#59a8d9'

color_converter = colors.ColorConverter().to_rgb
colormap = make_colormap([color_converter(c1), color_converter(c2), 0.6,
                          color_converter(c2), color_converter(c3), 0.8,
                          color_converter(c3)])

bgcolor = '#222222'
```

Plot Ross Lake, WA - the area I did my undergraduate thesis research in:

```python
rm = RidgeMap((-121.190529,48.773614,-121.018867,48.864941))
values = rm.get_elevation_data(num_lines=180)
ridges = rm.plot_map(values=rm.preprocess(values=values,
                                          lake_flatness=1,
                                          water_ntile=0,
                                          vertical_ratio=400),
                     label='',
                     kind='elevation',
                     linewidth=1,
                     background_color=bgcolor,
                     line_color = colormap)
```

Using a default color map:

```python
rm = RidgeMap((-121.190529,48.773614,-121.018867,48.864941))
values = rm.get_elevation_data(num_lines=180)
ridges = rm.plot_map(values=rm.preprocess(values=values,
                                          lake_flatness=1,
                                          water_ntile=0,
                                          vertical_ratio=400),
                     label='',
                     kind='elevation',
                     linewidth=1,
                     background_color=bgcolor,
                     line_color = plt.get_cmap('viridis'))

plt.savefig('Ross-Lake-ridge-map.jpg', dpi=300, bbox_inches='tight')
```

Plot Samre, where I am doing my Ph.D. research:
